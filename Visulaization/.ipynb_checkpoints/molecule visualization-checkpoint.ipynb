So i want to check how are all the descriptor values differentiating with solubility values with scatter plot
i created a snippet that can differentiate all the columns with solubility column using scatterplot

```python
features_start_at = list(df.columns).index('MolWt')
feature_names = df.columns[features_start_at:]

fig, axs = plt.subplots(nrows=7, ncols=4, sharey=True, figsize=(12, 8), dpi=400)
axs = axs.flatten() # don't want to think about i/j
for i,n in enumerate(feature_names):
    ax = axs[i]
    ax.scatter(
        df[n], df.Solubility, 
        s = 6, alpha=0.4,
        color = f'C{i}') # add some color 
    if i % 4 == 0:
        ax.set_ylabel('Solubility')
    ax.set_xlabel(n)
# hide empty subplots
for i in range(len(feature_names), len(axs)):
    fig.delaxes(axs[i])
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.histplot(df['Solubility'], ax=ax, bins=50, alpha=.5)

ax.set_title("Distribution of Solubility scores", weight="bold", size=15)
plt.rc("axes", labelsize=15)

mean_sol = df['Solubility'].mean()
median_sol = df['Solubility'].median()
ax.axvline(mean_sol, lw=4, color="r", label="Mean sol")
ax.axvline(median_sol, lw=4, color="green", label="Median sol")

ax.legend();
```
